## Vector Error Correction Models

Recall the Granger Representation Theorem : 'If there is cointegration there is a error correction term'.

A Vector Error Correction Model, VECM, is simply a re-parameterisation of a VAR. The re-parameterisation allows us to analyse vectors of I(1) variables.

The procedure developed by **Johansen (1988)** relies heavily on the rank of a matrix and its characteristic roots.

Consider a VAR(1) model:
$$ X_t = A_1X_{t-1} + \varepsilon _t$$

where $X_T$ is a n x 1 vector of I(1) variables. (y,x,z ...) 

$A_1$ is a n x n coefficient matrix.

We **reparameterise* the VAR by differencing, subtracting $X_{t-1}$ 
$$ X_t - X_{t-1} = A_1 X_{t-1} - X_{t-1} + \varepsilon _t$$
$$ \Delta X_t = (A_1 - I) X_{t-1} + \varepsilon _t$$

or: $$ \Delta X_t = \Pi X_{t-1} + \varepsilon _t$$
where: square matrix Pi is square matrix A - Square matrix I.

We test for cointegration using $\Pi$ , making use of the rank of the matrix.

Recall for stationarity we require roots to be less than 1 and thus need: $$(A_1 - I) \neq 0$$ 

The rank of the matrix is the number of linearly independent (rows or columns) n.b row rank = column rank.

A set of vectors {v1,v2,...,v_n} in a vector space are said to be linearly independent if for:
$$a_1v_1 + a_2v_2 + ... +a_nv_n = 0$$
$$a_1 = a_2 = ... = a_n = 0 $$
If we can find $a_i \neq 0$ for any i then the set of vectors are Linearly Dependent. If all the coefficients are found to be zero then the vectors are Linearly Independent.

In terms of matrices if no row can be expressed as a linear combination of other rows the rows are linearly indpendent.

The rank can be found by the number of non-zero rows in row echelon form of the matrix. Where each leading non-zero element appears to the right of the lead value of the row above.


**Definitions**

$\rho(\Pi)$ denotes the rank of the matrix, the number of linearly independent rows, number of non-zero rows in row echelon form. 

If $\rho(\Pi) = 0$ this means that all rows are linearly dependent on each other, i.e the row echelon form of all rows are zero. There is no linear combination among the I(1) variables that is I(0). All the variables in the vector are unit root processes and there is no cointegration. No linear combination that is stationary.

Another way to think about this is all eigenvalues that satisfy this equation are 1 implying non stationarity



If $\rho(\Pi) = n$ all the variables in the vector are stationary I(0) processes, if the rank of the matrix equals the number of rows this means all rows are linearly independent. If PI has full rank, then all variables are already independent and stationary on their own, meaning that no linear combination of variables exhibit the non-stationary behaviour needed for cointegraiton.

Since row rank = column rank. There is no linear combination where the coefficients of each variable is not zero. Each variable in the system is independently moving without being related to any other variable in a long-run stationary manner.

If $ 0 < \rho(\Pi) < n$ this means the matrix is reduced rank. The rank of the matrix is equal to the number of non-zero characteristic roots, there are r non-zero roots. For a rank less than n this means that some columns are dependent, implying the existence of stationary linear combinations among the variables that move together in the long run.

Some columns are linearly dependent on each others. These dependent columns represent cointegrating relationships. 

For clarity: rank less than n means that not all columns of PI are linearly independent, Implying that variables are related to each other in the long run, there are stationary linear combinations of the variables. We have differenced, I(1) data when differenced becomes a stationary series as the X term is removed and shocks only have transitory effects. When after differencing we find a linear relationship between variables, this means ther is a **stationary linear combination of variables** meaning there are cointegrating relationships.

If rank(Pi) = 1 , there is one cointegrating vector, one linear combination of variables that moves together in the long run.

With r independent equations and n variables, there are n-r stochastc trends in the system. As the matrix will have n-r zero eigenvalues.

## Johansen Methodology

The number of distinct **Cointegrating Vectors** can be obtained by checking the significance of the characteristic roots of PI.

The rank of PI = the number of characteristic roots that differ from zero.

The further the estimated characteristic roots are from zero, the more negative is $$ln(1 - \lambda _i)$$

For lambda less than 1 the function will be a negative number.

As lambda approaches 1, the term 1 - lambda approaches 0 and the function appraches negative infinity.

The closer lambda is to 0 the closer the function is to 0.

Eigenvalues close to 0 indicate the system is stationary.

Eigenvalues close to 1 indicate non-stationarity and the presence of unit roots.

The rank of the cointegration matrix determines the number of cointegrating vectors in the system. If the eigen values are close to one, this implies that the system has a unit root and the series are non-stationary. 

For clarity:

an eigenvalue close to zero indicates stationarity.

as lamda approaches zero, 1- lambda approaches 1 , ln (1-lambda) approaches negative infinity. 

**To determine Cointegrating rank:**

We use either the Trace Statistic or Maximum Eigenvalue Statistic:
$$
\text{Trace Statistic} : \  \lambda _{trace}(r) = -T \sum_{i=r+1}^{N} \ln(1 - \hat{\lambda}_i)
$$

$$
\text{Max Eigenvalue Statistic} : \  \lambda _{max}(r, r+1) = -T \ln(1 - \hat{\lambda}_{r+1})
$$

The closer lamda is to zero , ln(1-lamda) approaches negative infinity, which makes the above two statistics larger.

We have a critical value which determines whether we reject the null hypothesis of stationarity. I.e if the test statistic if above the critical value lamda is closer to zero than 1 and we have stationarity. 

The critical values are taken from the asymptotic distribution. They are upper tail tests. Where you find values large enough to reject null.

Recall in an upper tail test, critical values mark a threshold beyond which the test statistic is considered so extreme it is highly unlikely the null hypothesis is true. The test statistic being large indicates the observed data is inconsistent with the null hypothesis.

$$\textbf{Trace Test:}$$

$$\text{Stage } r:$$
$$
H_0: \text{The number of cointegration vectors is } r.
$$
$$
H_1: \text{The number of cointegration vectors is greater than } r.
$$

$$\text{Stage } r+1:$$
$$
H_0: \text{The number of cointegration vectors is at most r /}  \rho(\Pi) \leq r.
$$
$$
H_1: \text{The number of cointegration vectors is greater than } r.
$$

$$\textbf{Maximum Eigenvalue Test:}$$

$$\text{Stage } r:$$
$$
H_0: \text{The number of cointegration vectors is } r.
$$
$$
H_1: \text{The number of cointegration vectors is } r + 1.
$$

$$\text{Stage } r+1:$$
$$
H_0: \text{The number of cointegration vectors is at most r + 1 } / \rho(\Pi) \leq r.
$$
$$
H_1: \text{The number of cointegration vectors is greater than } r + 1.
$$

Let's follow the Johansen methodology in python:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(4)

n = 1000

epsilon_y = np.random.randn(n)
epsilon_z = np.random.randn(n)

series_y = np.cumsum(epsilon_y) # random walk effectively permanent shocks
series_z = np.cumsum(epsilon_z)

plt.figure(figsize=(12, 6))
plt.plot(series_y, label="y", color='blue')
plt.plot(series_z, label="z", color='crimson')
plt.legend()
plt.show()

from statsmodels.tsa.stattools import adfuller

adf_results = {
    'y' : adfuller(series_y, autolag = 'BIC'),
    'z' : adfuller(series_z, autolag = 'BIC') 
}

for series , result in adf_results.items():
    print(f" Results for {series}:")
    print(f"    ADF Statistic : {result[0]}")
    print(f"    p-value : {result[1]}")
    print(f"    Number of lags used: {result[2]}")
    print(f"    Critical Values:")
    for key , value in result[4].items():
        print(f"    {key}: {value}")
    print("=" * 55)
 
 # recall is non-stationarity

diff_adf_results = {
    'diff_y' : adfuller(np.diff(series_y), autolag = 'BIC'),
    'diff_z' : adfuller(np.diff(series_z), autolag = 'BIC') 
}


for series , result in diff_adf_results.items():
    print(f" Results for {series}:")
    print(f"    ADF Statistic : {result[0]}")
    print(f"    p-value : {result[1]}")
    print(f"    Number of lags used: {result[2]}")
    print(f"    Critical Values:")
    for key , value in result[4].items():
        print(f"    {key}: {value}")
    print("=" * 55)
```

Both series are I(1)

```python
from statsmodels.tsa.vector_ar.vecm import coint_johansen

data = np.stack([series_y, series_z], axis = 1)
#same as
data1 = np.column_stack([series_y, series_z])

johansen_test = coint_johansen(data, det_order= -1, k_ar_diff= 1)

print("Johansen Cointegration Test Results:")
print("Trace Statistic:", johansen_test.lr1)  # Trace statistic
print("Critical Values (5%):", johansen_test.cvt[:, 1])
print("Maximum eigenvalue statistic:", johansen_test.lr2)  # Maximum eigenvalue statistic
print("Critical Values (5%):", johansen_test.cvm[:, 1])
```

We reject the null hypothesis that there are no cointegrating vectors, and we cannot reject the null that there is exactly one cointegrating vector. 

Therefore these two series are cointegrated and have one cointegrating vector, the linear combination of y and z is stationary.

```python
resid = pd.DataFrame(johansen_test.r0t).plot()
```

## The Granger Representation Theorem 

The Granger representation theorem states that if a set of non-stationary variables are cointegrated then they can be characterized as generated by an error correction mechanism.

We can represent a vector of cointegrated variables in a **Vector Error correction Model**

Consider the n dimensional VECM:

$$ \Delta x_t = \Pi x_{t-1} + \sum^{p-1}_{i=1} \Pi \Delta x_{t-i} + \varepsilon _t$$

If $\rho (\Pi) = r $ i.e the coefficient matrix is reduced rank, indicating there are r independent linear combination of variables, there are r cointegrating relationships and n - r stochastic trends. The Long run matrix PI can be factorised into two n x r matrices.

i.e $$ \Delta x_t = \alpha \beta ' x_{t-1} + \sum^{p-1}_{i=1} \Pi \Delta x_{t-i} + \varepsilon _t$$

where $$ \Pi = \alpha \beta '$$

which are two nxr matrices and Beta has been transposed in order to multiply, as matrix multiplication is not commutative and the dimensions of the matrix must be compatable for matrix multiplication.

**KEY**

1) $\alpha$ is the 'loading matrix' or the 'speed of adjustment matrix'
2) $\beta$ is the matrix of cointegrating vectors, the weights on variables that make the linear combination stationary.

We are estimating the error correction relationship for a vector of variables, hence this is a Vector Error Correction Model.


**Granger Causality**

Granger causality is conducted on I(0) variables, do not test for Granger causality in cointegrated models **and do not conduct a first differenced VAR when the variables are cointegrated.** 

Reason being that Granger Causality tests conducted on I(1) variables do not have the standard F - distribution.

## Factorisation of the VECM

Consider the 3 Variable VECM:

$$
\begin{bmatrix}
\Delta x_t \\
\Delta y_t \\
\Delta z_t
\end{bmatrix}
=
\begin{bmatrix}
\pi_{11} & \pi_{12} & \pi_{13} \\
\pi_{21} & \pi_{22} & \pi_{23} \\
\pi_{31} & \pi_{32} & \pi_{33}
\end{bmatrix}
\begin{bmatrix}
x_{t-1} \\
y_{t-1} \\
z_{t-1}
\end{bmatrix}
+
\begin{bmatrix}
e_{1,t} \\
e_{2,t} \\
e_{3,t}
\end{bmatrix}
$$

If p(Pi) = 1 this means we have 1 Cointegrating Equation, and we decompose $\Pi = \alpha \beta'$:

$$
\begin{bmatrix}
\Delta x_t \\
\Delta y_t \\
\Delta z_t
\end{bmatrix}
=
\begin{bmatrix}
\alpha_{11} \\
\alpha_{21} \\
\alpha_{31}
\end{bmatrix}
\begin{bmatrix}
\beta_{11} & \beta_{12} & \beta_{13}
\end{bmatrix}
\begin{bmatrix}
x_{t-1} \\
y_{t-1} \\
z_{t-1}
\end{bmatrix}
+
\begin{bmatrix}
e_{1,t} \\
e_{2,t} \\
e_{3,t}
\end{bmatrix}
$$



Where the alpha matrix is the loading matrix, speed of adjustment matrix , which describes the speed of adjustment to disequillibrium of the long run Equillibrium relationship.

i.e:

$$
\begin{bmatrix}
\Delta x_t \\
\Delta y_t \\
\Delta z_t
\end{bmatrix}
=
\begin{bmatrix}
\alpha _{11}\beta _{11} & \alpha_{11}\beta _{12} & \alpha_{11}\beta _{13}\\
\alpha _{21} \beta _{11} & \alpha_{21}\beta _{12} & \alpha_{21}\beta _{13}\\
\alpha _{31} \beta _{11} & \alpha_{31}\beta _{12} & \alpha_{31}\beta _{13}
\end{bmatrix}
\begin{bmatrix}
x_{t-1} \\
y_{t-1} \\
z_{t-1}
\end{bmatrix}
+
\begin{bmatrix}
e_{1,t} \\
e_{2,t} \\
e_{3,t}
\end{bmatrix}
$$

that is: 

$$
\begin{bmatrix}
\Delta x_t \\
\Delta y_t \\
\Delta z_t
\end{bmatrix}
=
\begin{bmatrix}
\alpha_{11} (\beta_{11} x_{t-1} + \beta_{12} y_{t-1} + \beta_{13} z_{t-1}) \\
\alpha_{21} (\beta_{11} x_{t-1} + \beta_{12} y_{t-1} + \beta_{13} z_{t-1}) \\
\alpha_{31} (\beta_{11} x_{t-1} + \beta_{12} y_{t-1} + \beta_{13} z_{t-1})
\end{bmatrix}
+
\begin{bmatrix}
e_{1,t} \\
e_{2,t} \\
e_{3,t}
\end{bmatrix}
$$

where $(\beta_{11} x_{t-1} + \beta_{12} y_{t-1} + \beta_{13} z_{t-1})$ is the cointegrating equation/ long run stationary linear relationship. It is the cointegrating relationship between the variables at time t-1. Capturing the long run equilibrium relationship.

Interpretation:

1) x_t adjusts at speed of $\alpha _{11}$ to error $e_{1,t}$ , error term associated with $\Delta x_t$, from the cointegrating equation.
2) y_t adjusts at speed of $\alpha _{21}$ to error $e_{2,t}$ , error term associated with $\Delta y_t$, from the cointegrating equation.
3) z_t adjusts at speed of $\alpha _{31}$ to error $e_{3,t}$ , error term associated with $\Delta z_t$, from the cointegrating equation.

Each error term represents unexplained variation/shock specific to the corresponding variable, for example if there is a sudden change in x_t, the error term captures the variation:
1) the shock causes $\Delta x_t$ to deviate from its expected value based upon x,y,z _t-1 
2) the shock e1t affects the first difference of xt

## Summary

Over four periods the system has adjusted to the initial shock, at each step, the variables x,y,z are moving towards their long-term equillibrium as described in the cointegrating relationship. Lets visualise the shock.

```python
x = [2, 4.25, 6.35, 9.374, 13.728]
y = [3, 4, 5.68, 8.0992, 11.582]
z = [4, 4.5 , 5.34, 6.5496, 8.29] 
t = ['t-1', 't', 't+1', 't+2', 't+3'] 

delta_x = np.diff(x)
delta_y = np.diff(y)
delta_z = np.diff(z)

t_diff = t[1:]


plt.figure(figsize=(12, 6))


plt.subplot(1, 2, 1)
plt.plot(t, x, label='x', marker='o', color='blue', linestyle='-', markersize=8)
plt.plot(t, y, label='y', marker='o', color='green', linestyle='-', markersize=8)
plt.plot(t, z, label='z', marker='o', color='red', linestyle='-', markersize=8)
plt.title('Variables Over Time', fontsize=14)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(t_diff, delta_x, label='x', marker='o', color='blue', linestyle='-', markersize=8)
plt.plot(t_diff, delta_y, label='y', marker='o', color='green', linestyle='-', markersize=8)
plt.plot(t_diff, delta_z, label='z', marker='o', color='red', linestyle='-', markersize=8)
plt.title('Changes in Variables Over Time (Error Correction)', fontsize=14)
plt.xlabel('Time', fontsize=12)
plt.ylabel('Change ()', fontsize=12)
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

We can see for delta x at time t the value is larger than time t+1. The larger change at time t reflects the initial shock to the system, the smaller change in t+1 shows the system is adjusting to the shock and the initial shock has been partially corrected.

Let's see how to implement a VECM in python:

Recall our cointegrated series y and z from earlier.

```python
plt.figure(figsize=(12, 6))
plt.plot(series_y, label="y", color='blue')
plt.plot(series_z, label="z", color='crimson')
plt.legend()
plt.show()
```

```python
data = np.column_stack([series_y, series_z])

from statsmodels.tsa.vector_ar.vecm import select_order

lag_order = select_order(data = data, deterministic = 'n', maxlags = 10)
lag_order.summary()
```

```python
print(lag_order)
```

```python
# Rather than manually following the Johansen method we can use

from statsmodels.tsa.vector_ar.vecm import select_coint_rank

rank_test = select_coint_rank(data, 0, 0, method = 'trace', signif= 0.10)
print(rank_test)
```

Each row shows one test with the null hypothesis: 'The cointegration rank is r_x' and alternative: 'The cointegration rank is greater than r_x and less than r_y'. 
I.e row 1 null rank = 0 alternative rank < 2 rejecta
    row 2 null rank = 1 alternative rank < 2 accept

Meaning there is one cointegraiton equation:

# Parameter Estimation

To fit a VECM to the data we first build a VECM object where we define:
1) The deterministic terms
2) The lag order
3) The cointegration rank

Once we ahve the VECM instance we call its fit method which returns a VECMresults object. When we use the summary method on the VECMresults object it lists all the models parameter estimates.

Recall for a two variable VECM we will have: 
$$
\begin{bmatrix}
\Delta y_t \\
\Delta z_t
\end{bmatrix}
=
\begin{bmatrix}
\alpha_{11} (\beta_{11} y_{t-1} + \beta_{12} z_{t-1}) \\
\alpha_{21} (\beta_{11} y_{t-1} + \beta_{12} z_{t-1})
\end{bmatrix}
+
\begin{bmatrix}
e_{1,t} \\
e_{2,t}
\end{bmatrix}
$$

```python
from statsmodels.tsa.vector_ar.vecm import VECM

model = VECM(data, deterministic= 'n', k_ar_diff= lag_order.aic, coint_rank= rank_test.rank) #information criteria chose 0 lag 

vecm_res = model.fit()

print(vecm_res.summary())
```

```python
vecm_res.alpha
```

$$\alpha _{11} = -4.96042597e-05 = 0.0000496042597$$
$$\alpha _{21} = 3.34595804e-03 = 0.00334595804$$

$$\beta _{11} = 1.0000 $$
$$\beta _{21} = 6.6953 $$

In this case the cointegrating vector is [1, 6.6953]' which are the weights placed on the variables that make the linear combination stationary.

Alpha are the speed of adjustment coefficients the loading matrix is [[-4.9 x 10^-5] , [3.3 x 10^-3]] meaning that y adjusts to disequillibrium/error from the cointegrating equation at rate/speed of alpha 11 and z adjusts to error at rate alpha 21.

For robustness we still require normal errors for OLS and no residual autocorrelation.

```python
print(vecm_res.test_normality().summary())

white_test = vecm_res.test_whiteness(nlags = 10)
print(white_test.summary())
```

### Deterministic Term decision from statsmodels documentation

deterministicstr {"n", "co", "ci", "lo", "li"}

"n" - no deterministic terms

"co" - constant outside the cointegration relation

"ci" - constant within the cointegration relation

"lo" - linear trend outside the cointegration relation

"li" - linear trend within the cointegration relation

N.b: taken from eviews video

If we assume there is no deterministic trends in data, in other words we assume none of the variables have trends, we choose 'n', no intercept in long run equation CE/Beta

If we assume there is an intercept inside the Cointegrating Equation, but not in the VAR  we choose 'ci'. We assume there is a mean value for the cointegrating equation

If we assume there is no trend in the cointegrating equation but a constant in the VAR model we we use 'co', This assumes the change in delta y is non-zero at any time.

If we assume a linear trend in the VAR , we use linear trend outside the cointegrating equation. 'lo'

If we assume there is a linear trend, we assume the cointegrating equation has a trend, i.e the long run relationship has a trend we use, 'li'

If we can combine these such as 'coli' for constant in var linear trend in CE. Or 'colo' trend and intercept in var but outside CE.

It is unlikely to have the first difference trending. But we may have a constant or trend in the cointegrating long run relationship.

We call a trend restricted when it is restricted to the cointegrating equation.


# Identification of Vector Error Correction Model

Identification in relation to VECM's has a different meaning than Identification of VAR models.

We place restrictions on the VECM in order to test economic theory. 

We aim to overidentify the model in order to obtain a chi squared statistic to see if the restriction upholds and thus the economic theory is supported.

We cannot do this by exactly identifying the model so we must over-identify. 

Exact Identifying restrictions = r - 1  r is rank
Over Identifying restrictions  > r -1 

We use a chisquared statistic with degrees of freedom equal to the number of restrictions. 

$$
\begin{bmatrix}
\Delta x_t \\
\Delta y_t \\
\Delta z_t
\end{bmatrix}
=
\begin{bmatrix}
\alpha_{11} \\

\alpha_{31}

\end{bmatrix}
\begin{bmatrix}
x_{t-1} \\
y_{t-1} \\
z_{t-1}
\end{bmatrix}
+
\begin{bmatrix}
e_{1,t} \\
e_{2,t} \\
e_{3,t}
\end{bmatrix}
$$

One possible identification strategy is: 

$$
\begin{bmatrix}

\Delta y_t \\
\Delta z_t
\end{bmatrix}
=
\begin{bmatrix}
\ast \\
\ast \\

\begin{bmatrix}

\end{bmatrix}

\end{bmatrix}
+
\begin{bmatrix}
e_{1,t} \\

e_{3,t}
\end{bmatrix}
$$

where * means left unrestricted

we have imposed two restrictions which is greater than the rank 2 > 0 and so it is overidentified.

note: normalisation 1 is not a restriction

Here based on Economic Theory we have placed restrictions:
1) -1 restriction = xt and yt have a 1:1 relationship
2) 0 restriction on alpha 31 = no adjustment by zt the other variables adjust to disequillibrium however zt has zero adjustment speed.

Then we test these theories through the chi-square statistic.

If we wanted to test each restriction individually we could remove one and still get a chisquare statistic as 1 restriction is still overidentified 1 > 0.

lets say we want to place the single restriction that xt and yt have a 1:1 relationship.

```python
np.random.seed(13)
data = pd.DataFrame({
    'x': np.random.randn(100).cumsum(),
    'y': np.random.randn(100).cumsum(),
    'z': np.random.randn(100).cumsum()
})

lag_order = select_order(data, maxlags=5)
vecm = VECM(data, k_ar_diff= lag_order.aic, deterministic="ci", coint_rank=1)

vecm_fit = vecm.fit()

print("Estimated beta (cointegrating vectors):")
print(vecm_fit.beta)
```
